```python
from pyspark.sql.functions import current_timestamp, lit, col, to_date, expr
```

```python
##file ingestion --order table ingestion
orders_raw = (spark.read.csv(
    "/Volumes/de_workspace26/ecommerce_badal/raw/orders.csv",
    header=True, inferSchema=True)
  .withColumn("_ingested_at", current_timestamp())
  .withColumn("_source_file", lit("orders.csv"))
)
orders_raw.write.format("delta").mode("overwrite").saveAsTable("de_workspace26.ecommerce_badal.orders")
print(f"bronze.orders: {orders_raw.count()} rows")
```

```python
##ingest customer
customers_raw = (spark.read.csv(
    "/Volumes/de_workspace26/ecommerce_badal/raw/customers.csv",
    header=True, inferSchema=True)
  .withColumn("_ingested_at", current_timestamp())
  .withColumn("_source_file", lit("customers.csv"))
)
customers_raw.write.format("delta").mode("overwrite").saveAsTable("de_workspace26.ecommerce_badal.customers")
print(f"bronze.customers: {customers_raw.count()} rows")
```

```python
##ingest products
products_raw = (spark.read.csv(
    "/Volumes/de_workspace26/ecommerce_badal/raw/products.csv",
    header=True, inferSchema=True)
  .withColumn("_ingested_at", current_timestamp())
  .withColumn("_source_file", lit("products.csv"))
)
products_raw.write.format("delta").mode("overwrite").saveAsTable("de_workspace26.ecommerce_badal.products")
print(f"bronze.products: {products_raw.count()} rows")
```

```python
orders_stream_df = (spark.readStream
  .format("cloudFiles")
  .option("cloudFiles.format", "csv")
  .option("cloudFiles.schemaLocation", "/Volumes/de_workspace26/ecommerce_badal/checkpoints/bronze_orders_stream_schema")
  .option("header", "true")
  .option("inferSchema", "true")
  .load(VOLUME_PATH)
)

(orders_stream_df
  .withColumn("_ingested_at", current_timestamp())
  .writeStream
  .format("delta")
  .option("checkpointLocation", "/Volumes/de_workspace26/ecommerce_badal/checkpoints/bronze_orders_stream")
  .outputMode("append")
  .table("de_workspace26.ecommerce_badal.orders_stream")
)
```

```python
# Run this first to reset
dbutils.fs.rm("/Volumes/de_workspace26/ecommerce_badal/checkpoints/bronze_orders_stream", recurse=True)
dbutils.fs.rm("/Volumes/de_workspace26/ecommerce_badal/checkpoints/bronze_orders_stream_schema", recurse=True)
```

```python
# Stop any running stream first
for s in spark.streams.active:
    s.stop()

# Load only orders CSVs specifically
orders_stream_df = (spark.read
  .format("csv")
  .option("header", "true")
  .option("inferSchema", "true")
  .load([
      VOLUME_PATH + "orders.csv",
      VOLUME_PATH + "orders_batch2.csv"
  ])
  .withColumn("_ingested_at", current_timestamp())
  .withColumn("_source_file", lit("autoloader_simulation"))
)

# Clear old table first to remove nulls
spark.sql("DROP TABLE IF EXISTS de_workspace26.ecommerce_badal.orders_stream")

(orders_stream_df
  .write
  .format("delta")
  .mode("overwrite")
  .saveAsTable("de_workspace26.ecommerce_badal.orders_stream")
)

print(f"Rows written: {orders_stream_df.count()}")
spark.table("de_workspace26.ecommerce_badal.orders_stream").display()
```
